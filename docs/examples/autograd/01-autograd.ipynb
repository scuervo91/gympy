```python
import autograd.numpy as np
from autograd import elementwise_grad as grad
import matplotlib.pyplot as plt
```

```python
def prime(x):
    return np.square(x) + 1

def second(x):
    return np.sin(x)
```

```python
x = np.linspace(-2, 2, 100)
y1 = prime(x)
y2 = second(y1)

fig, ax = plt.subplots()

ax.plot(x, y1, label='prime')
ax.plot(x, y2, label='second')
ax.legend()
```

```python
grad_sec = grad(second)
```

```python
g2 = grad_sec(y1)

x = np.linspace(-2, 2, 100)
y1 = prime(x)
y2 = second(y1)

fig, ax = plt.subplots()

ax.plot(x, y1, label='prime')
ax.plot(x, y2, label='second')
ax.plot(x, g2, label='second gradient')
ax.legend()
```
