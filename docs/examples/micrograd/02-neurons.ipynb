```python
from gympy.autograd import Neuron
```

```python
x = [2.0,3.0]

linear = Neuron(2)
out = linear(x)

out
```

```python
relu = Neuron(2,type='relu')
out_relu = relu(x)
out_relu
```

```python
tanh = Neuron(2,type='tanh')
out_tanh = tanh(x)
out_tanh
```

```python
sigmoid = Neuron(2,type='sigmoid')
out_sigmoid = sigmoid(x)
out_sigmoid
```

```python
out_sigmoid.draw_dot()
```
